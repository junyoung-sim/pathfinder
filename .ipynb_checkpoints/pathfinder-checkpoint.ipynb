```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# undirected edges
edges = [
    (0,1),(0,3),(0,6),(0,12),
    (1,2),(1,4),(1,9),(1,13),
    (2,6),(2,7),(3,4),(3,5),
    (3,10),(4,11),(4,12),(4,14),
    (5,6),(6,7),(7,8),(7,9),
    (8,9),(8,13),(10,11),(13,14)
]

graph = nx.Graph()
graph.add_edges_from(edges)
pos = nx.spring_layout(graph)
nx.draw_networkx_nodes(graph, pos)
nx.draw_networkx_edges(graph, pos)
nx.draw_networkx_labels(graph, pos)
plt.show()
```

```python
# define q-value table
# let row be (s) and col be (s')
# set undefined edges to -1
qt = np.full((15,15), -1)
for edge in edges:
    qt[edge[0]][edge[1]] = 0
    qt[edge[1]][edge[0]] = 0

DST = 7 # destination
for s in range(0, 15):
    if qt[s][DST] == 0:
        qt[s][DST] = 10

print(qt)
```
